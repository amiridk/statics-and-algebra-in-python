```python
import numpy as np 
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn
```

```python
p = [.1 , .41 ,0 ,.334 , .22]
q = [.11 , .45 ,.12 ,.344 , .21]
```

```python
def KL(p , q):
    result = np.sum(np.where(np.logical_and(p != 0 , q != 0), p * np.log(p/q) ,0))
    return result
```

```python
x = np.arange(-10 , 10 , .001)
len(x)
```

```python
p = stats.norm.pdf(x , 0 , 2)
q = stats.norm.pdf(x, 2, 3)
plt.plot(x, p, 'b', x, q, 'r')
```

```python
KL(p ,q)
```

```python
p = stats.norm.pdf(x , 0 , 2)
q = stats.norm.pdf(x , .1 , 2.5)

plt.plot(x, p, c='blue')
plt.plot(x, q, c='red')
```

```python
KL(p ,q)
```
