```python
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_normal = stats.norm.rvs(size = 100000 , loc = 0 , scale = 1)
sns.histplot(data_normal, bins = 100, kde = True, color = 'blue')
plt.show()
```

```python
data_bernouli = stats.bernoulli.rvs(size=1000,p=0.2)

print(data_bernouli)
```

```python
# data_binom = stats.binom.rvs(size= 1000 , p= .2 , n=100)
data_binom = np.random.binomial(n=100, p=.2, size=1000)
print(data_binom)
```

```python
sns.histplot(data_binom , bins=100 , kde=True , color='red')
```

```python
p = 1/6
data_multinomial = np.random.multinomial(n =6 , pvals=[p,p,p,p,p,p] , size=1000)
print(data_multinomial)
```

```python
icu = .3
surgery = .2
el = .5

data_multinomial = np.random.multinomial(n=1000,pvals = [icu, surgery, el], size =365)
print(data_multinomial)
```

```python
data_poison = np.random.poisson(lam = 7 , size = 1000)
print(data_poison)
```

```python
sns.histplot(data_poison ,kde= True ,bins=100)
```

```python
data_exponential = np.random.exponential(scale=2 , size=1000)
print(data_exponential)
```

```python
sns.histplot(data_exponential , bins = 100 , color='blue')
```
